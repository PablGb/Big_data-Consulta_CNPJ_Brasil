```python
import pandas as pd
import pyarrow
import glob
import gc

diretorio_final = r'C:\Users\mayma\Documents\Projetos\Big_data-Consulta_CNPJ_Brasil\Base_de_dados_parquet'

dicionario_colunas = {
    'Empresas': ['cnpj_basico', 'razao_social', 'natureza_juridica', 'qualificacao_responsavel', 'capital_social', 'porte_empresa', 'ente_federativo_responsavel'],
    'Estabelecimentos': ['cnpj_basico', 'cnpj_ordem', 'cnpj_dv', 'identificador_matriz_filial', 'nome_fantasia', 'situacao_cadastral', 'data_situacao_cadastral', 'motivo_situacao_cadastral', 'nome_cidade_exterior', 'pais', 'data_inicio_atividade', 'cnae_fiscal_principal', 'cnae_fiscal_secundaria', 'tipo_logradouro', 'logradouro', 'numero', 'complemento', 'bairro', 'cep', 'uf', 'municipio', 'ddd_1', 'telefone_1', 'ddd_2', 'telefone_2', 'ddd_fax', 'fax', 'correio_eletronico', 'situacao_especial', 'data_situacao_especial'],
    'Simples': ['cnpj_basico', 'opcao_simples', 'data_opcao_simples', 'data_exclusao_simples', 'opcao_mei', 'data_opcao_mei', 'data_exclusao_mei'],
    'Socios': ['cnpj_basico', 'identificador_socio', 'nome_socio_razao_social', 'cnpj_cpf_socio', 'qualificacao_socio', 'data_entrada_sociedade', 'pais', 'representante_legal', 'nome_representante', 'qualificacao_representante_legal', 'faixa_etaria'],
    'Municipios': ['codigo_municipio', 'nome_municipio'],
    'Paises': ['codigo_pais', 'nome_pais'],
    'Qualificacoes': ['codigo_qualificacao', 'nome_qualificacao'],
    'Naturezas': ['codigo_natureza', 'nome_natureza'],
    'Cnaes': ['codigo_cnae', 'nome_cnae'],
    'Motivos': ['codigo_motivo', 'descricao_motivo']
}

for caminho in arqs:
    nome_arq = os.path.basename(caminho).replace('.zip', '')
    tamanho_arq = os.path.getsize(caminho) / (1024**2)
    colunas_arq = None

    for chave, colunas in dicionario_colunas.items():
        if nome_arq.startswith(chave):
            colunas_arq = colunas
            break
        
    if not colunas_arq:
        continue
    
    print(f"Processando: {nome_arq} ({tamanho_arq:.2f} MB)")

    if tamanho_arq > 300:
        porcao = pd.read_csv(caminho, sep=';', encoding='iso-8859-1', header=None, 
                             names=colunas_arq, dtype=str, chunksize=500000)
        
        lote_num = 1
        for lotes in porcao:
            nome_final = f'{nome_arq}_{lote_num}.parquet'
            caminho_final = os.path.join(diretorio_final, nome_final)
            
            if not os.path.exists(caminho_final):
                lotes.to_parquet(caminho_final, index=False)
            
            lote_num += 1
            del lotes
            gc.collect()
    else:
        nome_final = f'{nome_arq}.parquet'
        caminho_final = os.path.join(diretorio_final, nome_final)

        if not os.path.exists(caminho_final):
            df_inteiro = pd.read_csv(caminho, sep=';', encoding='iso-8859-1', 
                                     header=None, names=colunas_arq, dtype=str)
            df_inteiro.to_parquet(caminho_final, index=False)
            del df_inteiro
            gc.collect()

print("\n--- Conversão concluída ---")
```

**3. Higienização para Discador Automático (Foco em Regex)**
O time de telemarketing da sua empresa está enfrentando problemas porque o sistema de discagem automática está travando. O motivo é que os números de telefone na base da Receita muitas vezes vêm com letras, espaços, traços ou símbolos bizarros.
* **Tarefa:** Pegue a base de Estabelecimentos e faça uma limpeza na coluna principal de telefones usando Expressões Regulares (Regex) para substituir **qualquer caractere que não seja um número** por um espaço vazio. Exiba no seu código as primeiras linhas da tabela antes e depois da limpeza para comprovar que restaram apenas dígitos numéricos, e exporte a base limpa como `telefones_higienizados.parquet`.

```python
contatos_ruins = []
contatos = []

for path in dados_estab:
    df = pd.read_parquet(path)
    colunas = ['ddd_1', 'telefone_1', 'ddd_2', 'telefone_2']
    verificador = df[colunas].apply(lambda x: x.str.contains(r'[^0-9]', na = False, regex = True))
    mascara = verificador.any(axis = 1)
    df_problemas = df[colunas][mascara]
    if not df_problemas.empty:
        contatos_ruins.append(df_problemas)
        for coluna in colunas:
            df_problemas[coluna] = df_problemas[coluna].str.replace(r'\D', '', regex = True)
        contatos.append(df_problemas)    
    del df
    del df_problemas
    gc.collect()

contatos_ruins
```

```python
contatos_ruins = []
contatos = []

for path in dados_estab:
    df = pd.read_parquet(path)
    colunas = ['ddd_1', 'telefone_1', 'ddd_2', 'telefone_2']
    
    verificador = df[colunas].apply(lambda x: x.str.contains(r'[^0-9]', na=False, regex=True))
    mascara = verificador.any(axis=1)
    
    # Criamos a fatia com os problemas
    df_problemas = df[colunas][mascara]
    
    if not df_problemas.empty:
        # SALVAMOS UMA CÓPIA DA VERSÃO RUIM (Original)
        contatos_ruins.append(df_problemas.copy()) 
        
        # Agora alteramos o df_problemas original
        for coluna in colunas:
            df_problemas[coluna] = df_problemas[coluna].str.replace(r'\D', '', regex=True)
            
        # SALVAMOS A VERSÃO QUE FOI EDITADA
        contatos.append(df_problemas)    
    
    del df
    # O del df_problemas aqui é seguro pois as listas guardam as referências/cópias
    gc.collect()
```
